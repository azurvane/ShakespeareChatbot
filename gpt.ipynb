# tokenizing the text

> converting the raw string to numerical input according to some encodign and decoding algo

there are two types of tokenzer algo
- charecter level: tokenize every charecter
- subword: tokenize part of words (not complete word)

```python
stoi = {ch: i for i, ch in enumerate(char)}
itos = {i: ch for i, ch in enumerate(char)}

encoder = lambda s: [stoi[c] for c in s]
decoder = lambda l: ''.join([itos[i] for i in l])
print(encoder("hello world"))
print(decoder(encoder("hello world")))
```
